```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_data = pd.read_csv(clinical_trial_data_to_load)
# Combine the data into a single dataset

clinical_data = pd.merge(mouse_drug_data,clinical_trial_data, on = "Mouse ID", how = "outer")
# Display the data table for preview

clinical_data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
clinical_group = clinical_data.groupby(["Drug", "Timepoint"])
print(type(clinical_group))

# Convert to DataFrame
clinical_group_df = pd.DataFrame(clinical_group["Tumor Volume (mm3)"].mean()).reset_index()
# Preview DataFrame
clinical_group_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
group_std = clinical_group["Tumor Volume (mm3)"].sem()
# Convert to DataFrame
group_std_df = pd.DataFrame(group_std)
group_std_df = group_std_df.reset_index()
# Preview DataFrame
group_std_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
drug_summary = clinical_group_df.pivot(index = "Timepoint", columns = "Drug", values = "Tumor Volume (mm3)")
# Preview that Reformatting worked
drug_summary
```

```python
# Generate the Plot (with Error Bars)

drug_format_list = [("Capomulin","o","red"),("Infubinol","^","blue"),
                    ("Ketapril","s","green"),("Placebo","d","black")]
for drug, marker, colors in drug_format_list:
    error_bar = group_std[drug]
    tumor_response_plot = plt.errorbar(drug_summary.index, drug_summary[drug], error_bar,
                                      fmt=marker,ls='--',color=colors,linewidth=0.5)
    
plt.legend(loc="best")
plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.grid()

# Save the Figure
plt.savefig("tumor_response_to_treatment.png")
```

## Metastatic Response to Treatment

```python

# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
metastatic_sites = clinical_group["Metastatic Sites"].mean()
# Convert to DataFrame
metastatic_sites_df = pd.DataFrame(metastatic_sites).reset_index()
# Preview DataFrame
metastatic_sites_df.head(10)
```

```python

# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
metastatic_std = clinical_group["Metastatic Sites"].sem()
# Convert to DataFrame
metastatic_std_df = pd.DataFrame(metastatic_std).reset_index()
# Preview DataFrame
metastatic_std_df.head(10)
```

```python
# Minor Data Munging to Re-Format the Data Frames
metastatic_summary = metastatic_sites_df.pivot(index = "Timepoint", columns = "Drug", values = "Metastatic Sites")
# Preview that Reformatting worked
metastatic_summary
```

```python

# Generate the Plot (with Error Bars)
drug_format_list = [("Capomulin","o","red"),("Infubinol","^","blue"),
                    ("Ketapril","s","green"),("Placebo","d","black")]
for drug, marker, colors in drug_format_list:
    error_bar = metastatic_std[drug]
    metastatic_spread_plot = plt.errorbar(metastatic_summary.index, metastatic_summary[drug], error_bar,
                                      fmt=marker,ls='--',color=colors,linewidth=0.5)
    
plt.legend(loc="best")
plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration(Days)")
plt.ylabel("Met. Sites")
plt.grid()

# Save the Figure
plt.savefig("metastatic_spread_during_treatment.png")
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mice_count = clinical_group["Mouse ID"].nunique()
# Convert to DataFrame
mice_count_df = pd.DataFrame(mice_count).reset_index()
mice_count_df = mice_count_df.rename(columns = {"Mouse ID": "Mouse Count"})
# Preview DataFrame
mice_count_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mice_count_summary = mice_count_df.pivot(index = "Timepoint", columns = "Drug", values = "Mouse Count")
# Preview the Data Frame
mice_count_summary.head()
```

```python
drug_format_list = [("Capomulin","o","red"),("Infubinol","^","blue"),
                    ("Ketapril","s","green"),("Placebo","d","black")]
for drug, marker, colors in drug_format_list:
    total_mouse = mice_count_summary[drug][0]
    survival_rate = mice_count_summary[drug]/total_mouse*100
    survival_rate_plot = plt.plot(mice_count_summary.index, survival_rate,
                                      marker=marker,ls='--',color=colors,linewidth=0.5)
    
plt.legend(loc="best")
plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
plt.grid()


# Generate the Plot (Accounting for percentages)

# Save the Figure
plt.savefig("survival_during_treatment.png")
# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
percent_changes = (drug_summary.iloc[-1]-drug_summary.iloc[0])/drug_summary.iloc[0]*100
# Display the data to confirm
percent_changes
```

```python
# Store all Relevant Percent Changes into a Tuple
percent_changes_tuple = tuple(percent_changes)

# Splice the data between passing and failing drugs
passing = percent_changes < 0

# Orient widths. Add labels, tick marks, etc. 
drug_list = ["Capomulin", "Infubinol", "Ketapril", "Placebo"]
percent_change_list = [percent_changes[drug] for drug in drug_list]
percent_change_plot = plt.bar(drug_list,percent_change_list,width=-1,align="edge",color=passing.map({True:"g",False:"r"}))

plt.ylim(-30,70)
plt.ylabel("% Tumor Volume Change")
plt.title("Tumor Change over 45 Day Treatment")
plt.grid()

# Use functions to label the percentages of changes
def autolabel(bars):
    for bar in bars:
        height = bar.get_height()
        if height > 0:
            label_position = 2
        else:
            label_position = -10
        plt.text(bar.get_x() + bar.get_width()/2., label_position,
                "%d" % int(height)+"%",color="white",
                ha="center", va="bottom")


# Call functions to implement the function calls
autolabel(percent_change_plot)

# Save the Figure
plt.savefig("tumor_change_over_45day_treatment.png")

# Show the Figure
plt.show()
```
